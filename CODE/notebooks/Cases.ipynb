In this notebook we'll look into the case data and try to visualise and interpret it.

```r
library(ggplot2)
library(reshape2)
```

```r
case.data <- read.table("../../DATA/EPI_DATA/cases_FMD.txt", header = TRUE)
```

```r
case.data
```

```r
case.data.dt <- reshape2::melt(case.data, id = c("Year"),
                       variable.name = "serotype", value.name = "cases")
```

```r
options(repr.plot.width = 8, repr.plot.height = 6)
```

```r
z <- ggplot(case.data.dt, aes(x = Year, y = cases, fill = serotype, colour = serotype)) +
    geom_line() +
    geom_point() +
    scale_y_log10("FMD cases", expand = c(0, 0)) +
    scale_colour_manual(values=c("cadetblue", "indianred", "darkolivegreen4")) +
    scale_fill_manual(values=c("cadetblue", "indianred", "darkolivegreen4")) +
    theme_bw() + 
NULL
z
```

```r
ggsave("~/Dropbox/FMDV_AMERICA/REPO/FIGURES/PLOTS/cases_per_serotype_per_time.pdf", plot = z)
```
