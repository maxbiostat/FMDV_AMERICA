```r
library(RBeast)
```

```r
A.glm.log <- read.table(file = "~/DUMP/fmdv/combined_serotype_A_GLM_avmvn.country.glm.out", header = TRUE)
names(A.glm.log) <- gsub("country.c", "glmC", names(A.glm.log))
names(A.glm.log) <- gsub("glmCoefIndicators", "coefIndicator", names(A.glm.log))
A.glm.log <- A.glm.log[, -(grep("Times", names(A.glm.log)))]
```

```r
O.glm.log <- read.table(file = "~/DUMP/fmdv/combined_serotype_O_GLM.country.glm.out", header = TRUE)
names(O.glm.log) <- gsub("country.c", "glmC", names(O.glm.log))
names(O.glm.log) <- gsub("glmCoefIndicators", "coefIndicator", names(O.glm.log))
O.glm.log <- O.glm.log[, -(grep("Times", names(O.glm.log)))]
```

```r
# predictor.names <- c("borders",
#                      "cattleTrade_1986.1995", "cattleTrade_1995.2004", "cattleTrade_2004.2013",
#                     "diff_seqs_O", "geodistance",
#                     "pigsTrade_1986.1995", "pigsTrade_1995.2004", "pigsTrade_2004.2013",
#                     "prod_seqs_O", "samples_origin", "samples_destination",
#                     "sheepTrade_1986.1995", "sheepTrade_1995.2004", "sheepTrade_2004.2013")
predictor.names <- c("Borders",
                     "Cattle trade 1986-1995", "Cattle trade 1995-2004", "Cattle trade 2004-2013",
                    "Difference # sequences", "Geodistance",
                    "Pigs trade 1986-1995", "Pigs trade 1995-2004", "Pigs trade 2004-2013",
                    "Product # sequences", " # samples origin", " # samples destination",
                    "Sheep trade 1986-1995", "Sheep trade 1995-2004", "Sheep trade 2004-2013")
```

```r
plot_glm(Names = predictor.names, Log = A.glm.log, intercept = FALSE,
         conditional = TRUE, export = FALSE)
```

```r
plot_glm(Names = predictor.names, Log = O.glm.log, intercept = FALSE,
         conditional = TRUE, export = FALSE)
```

```r
plot_glm(Names = predictor.names, Log = A.glm.log, intercept = FALSE,
         title = "Serotype A",
         file_name = "../../FIGURES/PLOTS/glm_A",
         conditional = TRUE, export = TRUE)
```
