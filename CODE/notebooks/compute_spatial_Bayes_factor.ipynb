```r
## take out warmup and select only the relevant columns
get_info <- function(x, burnin = .2){
    x_b <- x[round(nrow(x) * burnin):nrow(x), ]
    return(x_b[, grep("country.indicators.", names(x_b))])
}
```

```r
filtered.A <- get_info(Log.A)
```

```r
filtered.O <- get_info(Log.O)
```

```r
parse_filtered <- function(filt, q){
    raw <- colMeans(filt)
    rq <- q/(1-q)
    Names <- names(raw)
    Split <- strsplit(Names, "\\.")
    ans <- data.frame(
        FROM =  unlist(lapply(Split, function(x) x[3])),
        TO = unlist(lapply(Split, function(x) x[4])),
        prob = as.numeric(raw) 
    )
    ans$BF <- (ans$prob /(1 - ans$prob) ) / rq
    ans$BF <- ifelse(is.finite(ans$BF), ans$BF, 1000)        
  return(ans)  
}
```

```r
KA <- 8
qA <- ( 7 + log(2) )/ ((KA * (KA - 1)) / 2)
qA
```

```r
KO <- 9
qO <- ( 8 + log(2) )/ ((KO * (KO - 1)) / 2)
qO
```

```r
( BFs.A <- parse_filtered(filtered.A, q = qA) )
```

```r
( BFs.O <- parse_filtered(filtered.O, q = qO) )
```

```r
sorted.BFs.A <- BFs.A[order(BFs.A$BF, decreasing = TRUE), ]
write.csv(sorted.BFs.A, "../../DATA/BEAST_OUTPUT/Bayes_factors_spatial_A.csv", row.names = FALSE)
```

```r
sorted.BFs.O <- BFs.O[order(BFs.O$BF, decreasing = TRUE), ]
write.csv(sorted.BFs.O, "../../DATA/BEAST_OUTPUT/Bayes_factors_spatial_O.csv", row.names = FALSE)
```
